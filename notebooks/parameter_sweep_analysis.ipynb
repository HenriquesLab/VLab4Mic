## VLab4Mic allows you to run parameter sweeps to explore the sample and imaging space!

```python
from vlab4mic import sweep_generator
import numpy as np
```

```python
sweep_gen = sweep_generator.sweep_generator()
```

# Set the parameters

```python
sweep_gen.structures = ["1XI5", ]
sweep_gen.probes = ["NHS_ester", ]
sweep_gen.modalities = ["STED", "Confocal"]
```

# Add the range and number of parameters to explore per group and type of parameter

```python
sweep_gen.set_parameter_values("probe", "labelling_efficiency", values=(0.5,1,2))
sweep_gen.set_parameter_values("virtual_sample", "random_orientations", values=[True,False])
sweep_gen.set_parameter_values("acquisition", "exp_time", values=(0.001,0.01,2))
```

# Analysis
### Set the reference to use for comparison

```python
sweep_gen.set_reference_parameters()
sweep_gen.generate_reference_image()
sweep_gen.preview_reference_image()
```

```python
sweep_gen.set_analysis_parameters(metrics_list=["ssim", "pearson"])
```

# Run the analysis

```python
sweep_gen.run_analysis(plots=True, save=True)
```
