```python
from supramolsim import sweep_generator
import numpy as np
```

```python
structures = ["1XI5", ]
probes = ["NHS_ester", ]
modalities = ["STED", "Confocal"]
```

```python
sweep_gen = sweep_generator.sweep_generator()
```

```python
sweep_gen.set_param_range("probe", "labelling_efficiency", "numeric", first=0.5, last=1, option=2)
sweep_gen.set_param_range("probe", "probe_distance_to_epitope", "numeric", first=0, last=100, option=2)
sweep_gen.set_param_range("virtual_sample", "random_orientations", "logical", option="True")
sweep_gen.set_param_range("acquisition", "exp_time", "numeric", first= 0.001, last = 0.1, option= 2)
```

```python
# automatically creates the itearables for all parameters, 
# then creates all virtual samples 
# and finally generates the acquisitions for each sample
sweep_gen.generate_acquisitions()
```

```python
sweep_gen.set_reference_parameters()
sweep_gen.generate_reference_image()
```

```python
sweep_gen.run_analysis()
sweep_gen.gen_analysis_dataframe()
```
